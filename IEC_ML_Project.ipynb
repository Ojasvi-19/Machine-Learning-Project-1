## Data Cleaning and Preprocessing

```python
#To check if we are having any missing value in our dataframe or not
df.isnull()
```

```python
#No Null Value in this dataframe
df.isnull().sum()
```

```python
df.duplicated()
```

```python
#No duplicates in this dataframe
df.duplicated().sum()
```

```python
df.dtypes
```

```python
#Change the datatype of the given column
df['Duration_Years']=df['Duration_Years'].astype('float32')
```

```python
df['Tuition_USD']=df['Tuition_USD'].astype('int32')
```

```python
df['Rent_USD']=df['Rent_USD'].astype('int32')
df['Visa_Fee_USD']=df['Visa_Fee_USD'].astype('int32')
df['Insurance_USD']=df['Insurance_USD'].astype('int32')
```
