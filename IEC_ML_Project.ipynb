##  Feature Encoding
Converting Categorical data into Numerical data because ML model only understands Numerical data.
There are 3 way to do this:
1. Label Encoder -> Most Primitive Technique   (Used in this project)
2. Ordinal Encoder -> Used for column having data in some order like Good, Better,Best.
3. One Hot Encoder

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le=LabelEncoder()
```

```python
df['Country']=le.fit_transform(df['Country'])
```

```python
df.head()
```

```python
df['City']=le.fit_transform(df['City'])
```

```python
df['University']=le.fit_transform(df['University'])
```

```python
df['Program']=le.fit_transform(df['Program'])
```

```python
df['Level']=le.fit_transform(df['Level'])
```

## Seperating Input and Output Data
Input Data (Raw Features): Original dataset columns.            
Output Data (Target/Label): The value you want to predict.

```python
X=df.iloc[:,:9]          #Traverse all Columns excluding index 9 i.e., Columns before index 9 -> 0-8
Y=df['Total Cost']
```

```python
X        #Input Data
```

```python
Y            #output Data -> Total cost column
```

## Feature Scaling

Make sure that all of the values of all of the columns of dataset must be in a range of 0-1. 

Scale all column values in a similar range.
There are 2 ways of doing so:
1. Standardization --> mean=0 & Standard Deviation=1 --> Used in ML
2. Normalization  --> Used in Neural Networks

```python
from sklearn.preprocessing import StandardScaler
s=StandardScaler()
```

```python
X=s.fit_transform(X)              #During Scaling one can directly pass on complete dataframe ata  once unlike in Feature encoding which needs to be done column by column
```

```python
X
```

Cannot scale output data as we want accurate results

## Train_Test_Split

We split both the input data (X) and output data (Y) into 80% training data and 20% testing data. The model is trained using X_train and Y_train. 
During testing, X_test is given to the trained model to generate predictions (Y_pred), which are then compared with the actual Y_test to evaluate the model's performance.

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=42)   #random state helps give exactly same test-train data whenever the program is executed, else whenever the program is run in its absence everytime aa new data is assigned.
```

## 3. Stacking

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import StackingRegressor
from sklearn.ensemble import BaggingRegressor
```

```python
estimators=[('rf',RandomForestRegressor()),('Br',BaggingRegressor())]
stack=StackingRegressor(estimators=estimators,final_estimator=DecisionTreeRegressor())
```

```python
stack.fit(X_train,Y_train)
```

```python
y_predict=stack.predict(X_test)
```

```python
mean_absolute_error(Y_test,y_predict)
```
